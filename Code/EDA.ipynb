# Loading the text file

```python
def load_doc(filename):
  # open the file as read only
  file = open(filename, 'r')
  # read all text
  text = file.read()
  # close the file
  file.close()
  return text
```

```python
filename = '../Datasets/Text.txt'
text = load_doc(filename)
text[0:1000]
```

```python
sentences = text.split('Page')
sentences[0]
```

# Cleaning the sentences

```python
# Taking only 200 pages of data for now i.e from Page 1 to Page 210
sentence = " ".join(sentences[1:201])
sentence
```

```python
cleaned_sentence = sentence.replace('\n',' ')
cleaned_sentence = cleaned_sentence.replace('\'', '')
cleaned_sentence = re.sub('[^a-zA-Z]', ' ', cleaned_sentence)
cleaned_sentence = re.sub(' +', " ", cleaned_sentence)
cleaned_sentence = cleaned_sentence.lower()
cleaned_sentence.strip()
```

```python
tokens = cleaned_sentence.split()
tokens[0:10]
```

```python
print('Number of tokens: {}'.format(len(tokens)))
print('Number of unique tokens: {}'.format(len(set(tokens))))
```

# Organising Sequences 

```python
#Organizing sequences such that each sequence or sentence has 51 words.
length = 50 + 1
sequences = list()

for i in range(length, len(tokens)+1):
  seq = tokens[i-length:i]
  line = ' '.join(seq)
  sequences.append(line)
print('Total Sequences: %d' % len(sequences))
```

# Saving the organized sequences

```python
# save tokens to file, one dialog per line
def save_doc(lines, filename):
  data = '\n'.join(lines)
  file = open(filename, 'w')
  file.write(data)
  file.close()
```

```python
# save sequences to file
out_filename = 'organized_sequences.txt'
save_doc(sequences, out_filename)
```
